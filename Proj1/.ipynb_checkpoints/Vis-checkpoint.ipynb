```python
def convert_to_np(x):
    return x.numpy()
```

```python
train_accuracy = list(map(convert_to_np, torch.load('train_accuracy.pt')))
train_loss = list(map(convert_to_np, torch.load('train_loss.pt')))
test_accuracy = list(map(convert_to_np, torch.load('test_accuracy.pt')))
test_loss = list(map(convert_to_np, torch.load('test_loss.pt')))
train_time = list(map(convert_to_np, torch.load('train_time.pt')))
```

```python
def get_stat(x):
    mean_x = np.mean(x, axis=0)
    std_x = np.std(x, axis=0)
    return mean_x, std_x
```

```python
epoch = np.arange(25)+1
```

```python
for acc, loss in zip(test_accuracy, test_loss):
    plt.figure(figsize=(20, 8))
    plt.subplot(1, 2, 1)
    acc_temp_1, acc_temp_2 = get_stat(acc)
    loss_temp_1, loss_temp_2 = get_stat(loss)
    plt.plot(epoch, acc_temp_1, c = 'gray')
    plt.fill_between(epoch, acc_temp_1 - acc_temp_2, acc_temp_1 + acc_temp_2, color = 'gray', alpha = 0.2)
    plt.subplot(1, 2, 2)
    plt.plot(epoch, loss_temp_1, c = 'gray')
    plt.fill_between(epoch, loss_temp_1 - loss_temp_2, loss_temp_1 + loss_temp_2, color = 'gray', alpha = 0.2)
```
